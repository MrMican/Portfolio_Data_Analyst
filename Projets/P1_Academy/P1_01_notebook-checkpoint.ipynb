## Creation de la DataFrame PIB

```python
#df_PIB correspond aux PIB par habitant en $ US pour chaque pays.

df_PIB=df_statsData.loc[df_statsData['Indicator Name']=="GDP per capita (current US$)"]
```

```python
#Application de la liste sur la df_PIB
#Classement dans l'ordre décroissant.

df_PIB=df_PIB[colonnes].dropna()
df_PIB.sort_values(['2014'], ascending=False)
```

Tranformation de l'indicateur **GPD en pourcentage**, cela me facilitera par la suite les calcules entre les différents colonnes

```python
#df_MaxPIB correspond au PIB maximum de l'année 2014, cette DataFrame me sera utilise par la suite pour passer mon indicteur GPD en poucentage.

df_MaxPIB=df_PIB['2014'].max()
df_MaxPIB
```

```python
#Passage de l'indicateur du GDP en pourcentage

df_PIB=df_PIB.assign(Percentage=lambda x :(x['2014']/df_MaxPIB*100))
df_PIB.sort_values(['2014'], ascending=False)            
```

Ici je crée une nouvelle **DataFrame PIB** qui regroupe seulement les données pour **les 5 pays à fort potentiel.**

```python
#df_SPIB correspond au classement des pays sélectionné en fonction de leur PIB.

df_SPIB=df_PIB.loc[df_PIB['Country Name'].isin(['Greece','Korea, Rep.','Australia','Spain','Belarus']) & (df_statsData['Indicator Name']=="GDP per capita (current US$)"),:]
df_SPIB[['Country Name','2014','Percentage']].sort_values(['Percentage'], ascending=False)
```

## Creation de la DataFrame Utilisateur d'internet 

```python
#df_Net correspond au pourcentage d'utilisateur d'internet pour 100 habitants pour chaque pays.

df_Net=df_statsData.loc[df_statsData['Indicator Name']=="Internet users (per 100 people)"]
```

```python
#Application de la liste sur la df_Net
#Classement dans l'ordre décroissant.

df_Net=df_Net[colonnes].dropna()
df_Net.sort_values(by=['2014'], ascending=False)
```

Ici je crée une nouvelle **DataFrame utilisateur d'internet** qui regroupe seulement les données pour **les 5 pays à fort potentiel.** 

```python
#df_SNet correspond au classement des pays sélectionné en fonction du pourcentage d'utilisateur.

df_SNet=df_statsData.loc[df_statsData['Country Name'].isin(['Greece','Korea, Rep.','Australia','Spain','Belarus']) & (df_statsData['Indicator Name']=="Internet users (per 100 people)"),:]
df_SNet[colonnes].sort_values(['2014'], ascending=False)
```

# Représentation Graphique et Sélection finale 

Voici deux courbes représentant **la DataFrame Utilisateur d'internet** en bleu et **la DataFrame PIB** en orange.

```python
x = df_SNet['Country Name']
y1 = df_SNet['2014']
y2 = df_SPIB['Percentage']

plt.plot(x, y1, label='Internet users %')
plt.plot(x, y2, label='GDP per capita %')
plt.title('Courbe des différentes Data Frame')
plt.xlabel('Country select')
plt.legend()
```

Je vais maintenant créer un **score** qui va nous permettre de voir plus clairement le pays avec le maximum de potentiel

```python
#Jointure entre la df_paysPIB et la df_paysNet.

df_final = df_SPIB.merge(df_SNet, left_on='Country Name', right_on='Country Name')

#Création de la df_final avec le score pour chaque pays.

df_final['Score'] = (df_final['2014_y'] + df_final['Percentage'])/2
df_final[['Country Name','2014_y','Percentage','Score']].sort_values(['Score'], ascending=False)
```

```python
#Création d'un histogramme qui permet de visualiser la df_final.

pays = df_final['Country Name']
score = df_final['Score']

plt.bar(pays, score)
plt.title('Score')
plt.xlabel('Country select')
plt.ylabel('Score en %')
```

Cet histogramme qui s'appuie sur **le score** de chaque pays nous permet de voir très clairement qu'un pays se démarque, l'**Australie** avec un score de **59%** suivie de la **Corée du sud 51%** et **l'Espagne 46%**.

```python
#Somme des taux d'utilisateurs d'internet et du PIB pour chaque pays.

largeur_barre = 0.8

y1 = df_final['2014_y']
y2 = df_final ['Percentage']
x = df_final['Country Name'] # position en abscisse des barres

# Tracé

plt.bar(x, y1, width = largeur_barre, color = "#3ED8C9")
plt.bar(x, y2, width = largeur_barre, bottom = y1, color = "#EDFF91")

plt.title('Somme')
plt.xlabel('Country select')
plt.ylabel('Score en %')
plt.xticks(df_final['Country Name']) ; plt.show()
```

Cet histogramme quant à lui s'appuie sur le somme des valeurs de la Data Frame final.
