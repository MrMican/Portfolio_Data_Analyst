## Elimination des prix négatifs

```python
df_products.describe()
```

```python
df_products = df_products.drop(df_products[df_products.price < 0].index)
df_products
df_products.describe()
```

```python
df_products.describe(include='all')
```

## Elimination des valeurs "tests"

```python
df_transactions.describe(include='all')
```

```python
df_transactions= df_transactions[df_transactions['date'].str.contains('test') == False]
df_transactions.describe(include='all')
```

## Jointure des DataFrames 

```python
df_j= pandas.merge(df_customers, df_transactions, how = 'outer')
df_j
```

```python
df_j = pandas.merge(df_j, df_products,how = 'outer' )
df_j
```

## Imputation des valeurs manquantes

```python
df_j.isnull().sum()
```

```python
df_j= df_j[df_j['session_id'].isnull() == False]
df_j
```

```python
df_j.isnull().sum()
```

```python
df_j[df_j.isna().any(axis=1)]
```

```python
df_j['price'].fillna(df_products['price'].mean(), inplace = True)
df_j['categ'].fillna(df_products['categ'].mode()[0], inplace = True)
df_j
```

```python
df_j.isnull().sum()
```

```python
df_j.to_csv('DataClean.csv', index=False)  
```
